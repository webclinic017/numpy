Interpolation is used to fill in the missing data between known data points. However, you should be careful if the data you have is a suitable for interpolation.

Numpy provides the `interp1d()` to do linear or cubic interpolation. 

````python
import numpy as np
````

Travel class and fare

````python
class_travel = [1, 3, 4]
fare_travel = [100, 60, 20]
````

Interpolate the fare for the 2nd class.

````python
np.interp(2, class_travel, fare_travel)
````

The `class_travel` is treated as the index here. It is based on these values the `fare_travel` is interpolated

__Important__: Make sure the data is sorted based on the index array before you interpolate.



[Image source](https://ncalculators.com/geometry/linear-interpolation-calculator.htm)

What if there were intermediate classes?


````python
np.interp([1.5, 2, 2.5, 2.9], class_travel, fare_travel)
````

<div class="alert alert-info" style="background-color:#006666; color:white; padding:0px 10px; border-radius:5px;"><h2 style='margin:7px 5px; font-size:16px'>Sine Wave Interpolation</h2>
</div>

````python
# turn off scientific notation
np.set_printoptions(suppress=True)
````

First generate 25 points of sine wave. Then interpolate the intermediate points linearly.

````python
x = np.linspace(0, 4*np.pi, 25)
y = np.sin(x)
````

````python
# Plot
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(x, y, 'ro-')
plt.show()
````

````python
# Interpolate
xvals = np.linspace(0, 4*np.pi, 100)
yinterp = np.interp(xvals, x, y)
yinterp
````

````python
# Plot
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(x, y, 'ro')
plt.plot(xvals, yinterp, '-x')
plt.show()
````

<div class="alert alert-info" style="background-color:#006666; color:white; padding:0px 10px; border-radius:5px;"><h2 style='margin:7px 5px; font-size:16px'>Iterpolate with SciPy</h2>
</div>

````python
# Import Package
# !pip install scipy
````

````python
from scipy import interpolate
````

````python
fx = interpolate.interp1d(x, y)
fx
````

````python
xvals = np.linspace(0, 4*np.pi, 100)
ynew = fx(xvals)
ynew
````

````python
# Plot
plt.plot(x, y, 'ro')
plt.plot(xvals, ynew, '-x')
plt.show()
````

__Cubic Interpolation is more appropriate in this case__

````python
fx = interpolate.interp1d(x, y, kind="cubic")
fx
````

````python
ynew = fx(xvals)
ynew
````

````python
# More smoother curve
plt.plot(x, y, 'ro')
plt.plot(xvals, ynew, '-x')
plt.show()
````

__Solution:__

````python
# 1.
import numpy as np
arr = np.genfromtxt("Datasets/Class_and_Fare.csv", delimiter=",", skip_header=1)
arr[:10]
````

````python
# Create the inputs to np.interp
x = arr[np.isnan(arr[:, 1]), 0]
xs = arr[~np.isnan(arr[:, 1]), 0]
ys = arr[~np.isnan(arr[:, 1]), 1]
````

````python
# Sort x and Y before interpolating
loc = np.argsort(xs)
xs = xs[loc]
ys = ys[loc]
````

````python
# interpolate
y = np.interp(x, xs, ys)
y
````

````python
# result
np.c_[x, y]
````

````python
# 2.
from scipy import interpolate
x = np.linspace(-18, 18, 36)
noise = 0.5 * np.random.random(len(x))
signal = np.sinc(x) + noise

x_input = np.linspace(-18, 18, 180)
````

````python
# linear
model_linear = interpolate.interp1d(x, signal, kind="linear")
y_linear = model_linear(x_input)
y_linear[:10]            
````

````python
# cubic
model_cubic = interpolate.interp1d(x, signal, kind="cubic")
y_cubic = model_cubic(x_input)
y_cubic[:10]            
````

__Plot__

````python
# Plot
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
plt.figure(figsize=(12,7))
plt.plot(x, signal, 'o', label="data")
plt.plot(x_input, y_linear, 'r--', label="linear")
plt.plot(x_input, y_cubic, '-', label="cubic")


plt.title("Interpolation: Cubic vs Linear", fontdict={'size': 16})
plt.legend()
plt.show()
````
