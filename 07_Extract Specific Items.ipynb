````python
data[:, 3] == np.nan
````

__Rows that contain missing in any of the columns__

The `any` method will return `True` if any of the values is true. Setting `axis=1` will do it row-wise. `axis=0` is column wise.

````python
data[np.isnan(data).any(axis=1), :]
````

````python
np.isnan(data).any(axis=1)
````

````python
~np.isnan(data).any(axis=1)
````

__Drop all rows that contain one or more missing__

````python
data[~np.isnan(data).any(axis=1), :]
````

__Get the maximum value in each row__

`axis=1` will do it row-wise.

````python
data.max(axis=1)
````

_If a row contains missing values, it returns `nan`. So use `np.nanmax` instead to ignore missing._

````python
np.nanmax(data, axis=1)
````

Equivalently, you have `nanmin`, `nanmean`, `nanmedian`, `nanpercentile` functions.

__Get the maximum value in each column__

````python
data.max(axis=0)
````

Ignore missing data with `nanmax`.

````python
np.nanmax(data, axis=0)
````

__Writing if-else logic using `np.where()`__

If-Else logic using `np.where()`.

Example logic: If the second column = 1, then, keep score (5th col) as is, else divide it by 2.

````python
np.where(data[:, 1] == 1, data[:, 4], data[:, 4]/2)
````

__Get the position of the maximum value in each row__

````python
max_pos = np.argmax(data, axis=1)
max_pos
````

````python
data[:5]
````

````python
# max values in each row.
[data[row, i] for row, i in enumerate(max_pos)]
````

__Get the position of values that satisfy a given condition__

````python
# Positions where a given condition is satisfied
pos = np.argwhere(data[:, 1] == 1)
pos[:5]
````

1. From `Mall_Customers_Int.csv`, find the row positions where 2nd column is 1 and 3rd column has value < 21. Extract the values from these columns. How many such rows exist?

2. Create a new array from `Mall_Customers_Int.csv` that has the value 1 if 2nd column = 1 and 3rd column < 21. Otherwise, it has the value 0.

```python
import numpy as np
data = np.genfromtxt('Datasets/Mall_Customers_Int.csv', 
                     delimiter=",",
                     skip_header=1)
```

__Solutions__

````python
import numpy as np
data = np.genfromtxt('Datasets/Mall_Customers_Int.csv', 
                     delimiter=",",
                     skip_header=1)
````

````python
# Solution 1
pos = np.argwhere((data[:, 1] == 1) & (data[:, 2] < 21))
pos[:5]
````

````python
# how many such exist?
len(pos)
````

````python
# Solution 2
arr = np.where((data[:, 1] == 1) & (data[:, 2] < 21), 1, 0)
arr
````

````python
# Check
data[:5, :]
````
